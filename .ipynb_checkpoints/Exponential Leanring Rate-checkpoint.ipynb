# Exploring Learning Exponential Learning Rate Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

## From Hands-on-ML 2ed

> ***Note:***  
> Using <code>steps</code> as a paramater makes it hard to predict the decay behaviour.  
> Hard to know the *final learning rate*.  
> And can't control rate of decay.

```python
lr_initial = 0.01
steps = 50
x = np.arange(1, 101, 1)
y = lr_initial * 0.1 ** (x/steps)
print(f'Final lr:{y[-1]}') 
plt.plot(x, y)
plt.show()
```

## Modified Version

> ***Note:***
> By controlling the <code>decay_rate</code> and fixing the number of steps to the number of epochs  
> The *final learning rate* equals: <code>lr_initial * decay_rate</code>  
> And the rate of decay is controlabel.

```python
lr_initial = 8.171579676021561e-05
decay_rate = 0.01
total_epochs = 100

x = np.arange(1, total_epochs + 1, 1)
y = lr_initial * decay_rate ** (x/total_epochs)
plt.plot(x, y)
plt.show()
print(f'Final lr:{y[-1]}') 
for i, lr in enumerate(y):
    print(f'epoch {i+1}: {lr}')
```
